```python
for num in range(100):
    print("num")
    if num % 2 == 0:
        continue # forces the loop to go to the next loop value
    print(num)
```

```python
for num in range(100):
    print("num")
    if num % 2 == 0:
        break # forces the loop stop completely
    else:
        print(num)
```
